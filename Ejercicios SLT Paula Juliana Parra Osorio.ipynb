1) El algoritmo perceptrón es un algoritmo iterativo que se basa en la siguiente regla de actualización simple:

Si
\begin{array}{lcc}
y(x*w) \leq 0;\\
w = w+xy
\end{array}

Donde y es la etiqueta (ya sea -1 o +1) de nuestro punto de datos actual x, y w es el vector de pesos.

El producto escalar x⋅w es solo la predicción del perceptrón basada en los pesos actuales (su signo es el mismo que el de la etiqueta predicha). La expresión y(x⋅w) puede ser menor o igual a 0 solo si la etiqueta real y es diferente de la etiqueta predicha $\varphi(x⋅w)$. Entonces, si hay una discrepancia entre las etiquetas verdaderas y predichas, entonces actualizamos nuestros pesos: w = w+yx; de lo contrario, los dejamos como están.

2) Sí, el algoritmo de aprendizaje del perceptrón es un clasificador lineal. Si sus datos son separables por un hiperplano, entonces el perceptrón siempre convergerá. Nunca convergerá si los datos no son linealmente separables.

3) El ángulo entre w y x debe ser menor que 90 porque el coseno del ángulo es proporcional al producto escalar. Sin importar cuánto es w, siempre que forme un ángulo de menos de 90 grados con los vectores de datos de ejemplo positivos (x E P) y un ángulo de más de 90 grados con los vectores de datos de ejemplo negativos (x E N), estamos bien. Entonces, el ángulo entre w y x debe ser menor a 90 cuando x pertenece a la clase P y el ángulo entre ellos debe ser mayor a 90 cuando x pertenece a la clase N. Cuando se suma x a w, lo que hacemos cuando x pertenece a P y w.x < 0, esencialmente estamos aumentando el valor cos(alfa), lo que significa que estamos disminuyendo el ángulo entre w y x, que es lo que deseamos. Y la intuición similar funciona para el caso en que x pertenece a N y w.x ≥ 0.

$h(x)=sign(w^Tx)$=  \begin{array}{lcc}
             +1, & w^Tx>0 \\

Así,

$y(t)h(x)=y(t)w^T(t)x(t)=−1 \rightarrow y(t)w^T(t)x(t)<0$.

$y(t)w^T(t+1)x(t)=y(t)w^T(t)x(t)+x^T(t)x(t)>y(t)w^T(t)x(t)$.

Here is an experiment that illustrates the difference between a single bin

Flip each coin independently times. Let's focus on 3 coins as follows:

* (a) What is µ for the three coins selected? 

µ = 0.5, que es la probabilidad esperada en una moneda balanceada.

* (b) Repeat this entire experiment a large number of times (e.g. , 100.000
runs of the entire experiment) to get several instances of v1 , Vrand
and Vmin and plot the histograms of the distributions of v1 , Vrand and
Vmin · Notice that which coins end up being Crand and Cmin may differ

```julia
#Cargue librerías
using Base
using StatsBase
using Plots
```

```julia
#Función para simular el lanzamiento de monedas (Cara = 1, Sello = 0)
function simular(n_monedas, n_veces)
    resultado = rand((0, 1),(n_monedas, n_veces))
    caras = sum(resultado, dims = 2)
    caras /= 10
    return caras[1], rand(caras), minimum(caras)
end
```

```julia
#Simulación de lanzamiento de 1.000 monedas 10 veces cada una suponiendo que están balanceadas. 
n_monedas = 1000;
n_veces = 10;
mu = 0.5;
experimentos = ones(100000, 3);

for i in 1:size(experimentos)[1]
    first, rand, minimun = simular(n_monedas, n_veces);
    experimentos[i, 1], experimentos[i, 2], experimentos[i, 3] = first, rand, minimun;
end
```

#### Simulación de la primera moneda

```julia
bins = LinRange(0, 1, 5)
plot(fit(Histogram, experimentos[:,1], bins))
```

#### Simulación de moneda aleatoria

```julia
plot(fit(Histogram, experimentos[:,2], bins))
```

#### Simulación de la moneda con menos caras

```julia
plot(fit(Histogram, experimentos[:,3], bins))
```

* (c) Using (b), plot estimates for P[|v-µ| > E] as a function of E, together
with the Hoeffding bound $2e^{-2e^2N}$ (on the same graph). 

```julia
#Función que genera la probabilidad de equivocarse. 
function prob(outcomes, mu, epsilon)
    less = sum(outcomes .< mu - epsilon)
    greater = sum(outcomes .> mu + epsilon)
    return ((less + greater) / size(outcomes)[1])
end

#Función para calcular la cota superior en la desigualdad de Hoeffding
function hoeffding(epsilon, N)
    return 2 * exp(-2 * epsilon * epsilon * N)
end
```

```julia
#Cálculo de la probabilidad y desigualdad de Hoeffding para varios valores de epsilon
epsilons = LinRange(0, 1, 100)
P_first = [prob(experimentos[:, 1], mu, epsilon) for epsilon in epsilons];
P_random = [prob(experimentos[:, 2], mu, epsilon) for epsilon in epsilons];
P_min = [prob(experimentos[:, 3], mu, epsilon) for epsilon in epsilons];
P_hoeffding = [hoeffding(epsilon, n_veces) for epsilon in epsilons];
```

```julia
plot(epsilons, [P_first, P_random, P_min, P_hoeffding], label = ["primero" "aleatorio" "mínimo" "Hoeffding"])
```
